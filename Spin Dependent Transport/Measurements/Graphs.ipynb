```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
for file in os.listdir():
    if file.endswith('.xlsx'):
        Data = pd.read_excel(file)
        Data.index.name = 'Time'
        N = Data.index[-1]
        AVG = sum(Data.iloc[:,1])/(N+1)
        SUM = sum(np.sqrt([i**2 for i in (Data.iloc[:,1]-AVG)]))
        Data = Data[Data.iloc[:,1]-AVG < 10*SUM/N]
        print(file)
        plt.scatter(Data.iloc[:,0],Data.iloc[:,1])
        plt.show()
```
